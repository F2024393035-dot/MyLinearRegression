```python
#Import Required Libraries
import numpy as np
```

```python
#Define Your Own Linear Regression Class
class MyLinearRegression:
    def __init__(self):
        self.m = 0
        self.b = 0

    #Implement the fit() Method
    def fit(self, x, y):
            x = np.array(x)
            y = np.array(y)

            # Mean calculations
            x_mean = np.mean(x)
            y_mean = np.mean(y)

            # Slope (m)
            numerator = np.sum((x - x_mean) * (y - y_mean))
            denominator = np.sum((x - x_mean) ** 2)
            self.m = numerator / denominator

            # Intercept (b)
            self.b = y_mean - self.m * x_mean

    #Implement the predict() Method
    def predict(self, x):
            x = np.array(x)
            return self.m * x + self.b

#Test Your Library
# Example data
x_train = [1, 2, 3, 4, 5]
y_train = [2, 4, 5, 4, 5]

# Create model
model = MyLinearRegression()

# Train (fit)
model.fit(x_train, y_train)

# Predict
predictions = model.predict([6, 7, 8])
print("Predicted values:", predictions)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_train, y_train, color='blue', label='Original Data')
plt.plot(x_train, model.predict(x_train), color='red', label='Fitted Line')
plt.legend()
plt.show()
```
